```python
# Cell 2: Load Model
from ultralytics import YOLO
model = YOLO("yolo11s.pt")
print("✅ Model loaded")
```

```python
# Cell 3: Custom Training (if you have your own dataset)
results = model.train(
    data="your_retail_dataset.yaml",   # Create on Roboflow
    epochs=50,
    imgsz=640,
    batch=16,
    name="retail_model",
    patience=20
)
```

```python
# Cell 4: Inference & Heatmap Generation
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def generate_customer_heatmap(frame, results):
    """Create heatmap of customer positions"""
    heatmap = np.zeros((frame.shape[0], frame.shape[1]), dtype=np.float32)
    
    for box in results[0].boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        center_x = int((x1 + x2) / 2)
        center_y = int((y1 + y2) / 2)
        heatmap[center_y-20:center_y+20, center_x-20:center_x+20] += 1
    
    plt.figure(figsize=(12, 8))
    sns.heatmap(heatmap, cmap="Reds", alpha=0.7)
    plt.title("Customer Density Heatmap")
    plt.axis('off')
    plt.savefig("customer_heatmap.png")
    plt.show()
```
